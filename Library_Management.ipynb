```python
# Library dictionary
library = {}

def add_book(book_id, name, author, quantity):
    library[book_id] = {
        "name": name,
        "author": author,
        "quantity": quantity
    }
    print("Book added successfully")

def delete_book(book_id):
    if book_id in library:
        del library[book_id]
        print("Book deleted successfully")
    else:
        print("Book ID not found")

def total_books():
    print("Total number of books:", len(library))

def display_books():
    if not library:
        print("No books available")
    else:
        for book_id, details in library.items():
            print("Book ID:", book_id)
            print("Name:", details["name"])
            print("Author:", details["author"])
            print("Quantity:", details["quantity"])
            print("----------------------")

def search_book(book_id):
    if book_id in library:
        details = library[book_id]
        print("Book Found")
        print("Name:", details["name"])
        print("Author:", details["author"])
        print("Quantity:", details["quantity"])
    else:
        print("Book not found")
```

```python
add_book(101, "Python", "Guido", 5)
add_book(102, "Java", "James", 3)
display_books()
total_books()
search_book(101)
delete_book(102)
display_books()

```
