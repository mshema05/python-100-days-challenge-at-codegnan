```python
#set not allows  list as element
s1={1,2,3,4,'hi',1,2,3,'a',[1,2,3,4]}
print(s1)
```

```python
#set not allows dictionary  as element
s1={1,2,3,4,'hi',1,2,3,'a',{1:2}}
print(s1)
```

```python
##set up funtions
#adding elements to set
#add(ele):it add the element to set if element is not in set otherwise nothing
#update(iterable): it add the all elements in iterator
```

```python
s1={1,2,3,4}
s1.add(10)
s1.add(1)
print (s1)
s1.update((5,3,5))
print(s1)
```

```python
s1={1,2,3,4}
s1.remove(1)
print(s1)
```

```python
##remove element from set
#remove(ele): it removes the elements from set if it is present in set otherwise keyerror
#discard(ele):it removes the elements from set if it is present in set otherwise nothing
#pop():it removes one random and return one element from set
s1={1,2,3,4}
s1.remove(3)
s1.discard(4)
print(s1)
print(s1.pop())
print(s1)
```

```python
s1={1,2,3,4,5,0,7}
s2={4,7,4,8,0}
a=s1.union(s2)#|
b=s1.intersection(s2)#&
c=s1.difference(s2)#-
d={1}.issubset(s1)
e=s1.symmetric_difference(s2)
print(a)
print(b)
print(c)
print(d)
print(e)
```
